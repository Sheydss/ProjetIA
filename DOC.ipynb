# **Documentation des champs de données**

Ce document décrit la signification de chaque champ dans le jeu de données des employés.

## **Données personnelles et démographiques**

### **1. Age**
- **Description :** Âge de l'employé en années.

### **2. Gender**
- **Description :** Genre de l'employé (`Male` ou `Female`).

### **3. MaritalStatus**
- **Description :** Statut marital de l'employé (`Single`, `Married`, ou `Divorced`).

## **Informations professionnelles**

### **4. Attrition**
- **Description :** Indique si l'employé a quitté l'entreprise (`Yes` ou `No`).

### **5. BusinessTravel**
- **Description :** Fréquence des déplacements professionnels (`Rarely`, `Frequently`, ou `Non-Travel`).

### **6. Department**
- **Description :** Département dans lequel l'employé travaille (`Sales`, `Research & Development`, ou `Human Resources`).

### **7. JobLevel**
- **Description :** Niveau hiérarchique de l'employé (valeur numérique).

### **8. JobRole**
- **Description :** Rôle ou poste spécifique de l'employé dans l'organisation (ex. `Manager`, `Research Scientist`).

### **9. MonthlyIncome**
- **Description :** Revenu mensuel de l'employé en dollars.

### **10. TotalWorkingYears**
- **Description :** Nombre total d'années pendant lesquelles l'employé a travaillé au cours de sa carrière.

### **11. YearsAtCompany**
- **Description :** Nombre d'années pendant lesquelles l'employé a travaillé dans l'entreprise actuelle.

### **12. YearsSinceLastPromotion**
- **Description :** Nombre d'années écoulées depuis la dernière promotion de l'employé.

### **13. YearsWithCurrManager**
- **Description :** Nombre d'années pendant lesquelles l'employé a travaillé sous la supervision de son manager actuel.

### **14. JobInvolvement**
- **Description :** Niveau d'implication de l'employé dans son travail (échelle de 1 à 4).

### **15. PerformanceRating**
- **Description :** Évaluation de la performance de l'employé (échelle de 1 à 4).

## **Formation et éducation**

### **16. Education**
- **Description :** Niveau d'éducation de l'employé (échelle de 1 à 5).

### **17. EducationField**
- **Description :** Domaine principal de formation de l'employé (ex. `Life Sciences`, `Marketing`).

### **18. TrainingTimesLastYear**
- **Description :** Nombre de fois où l'employé a suivi une formation au cours de l'année dernière.

## **Satisfaction et équilibre**

### **19. EnvironmentSatisfaction**
- **Description :** Niveau de satisfaction de l'employé vis-à-vis de son environnement de travail (échelle de 1 à 4).

### **20. JobSatisfaction**
- **Description :** Niveau de satisfaction de l'employé vis-à-vis de son poste (échelle de 1 à 4).

### **21. WorkLifeBalance**
- **Description :** Évaluation de l'équilibre travail-vie personnelle de l'employé (échelle de 1 à 4).

## **Informations diverses**

### **22. DistanceFromHome**
- **Description :** Distance entre le domicile de l'employé et le lieu de travail (en kilomètres).

### **23. NumCompaniesWorked**
- **Description :** Nombre d'entreprises pour lesquelles l'employé a travaillé avant l'actuelle.

### **24. Over18**
- **Description :** Indique si l'employé est majeur (valeur fixe `Yes`).

### **25. PercentSalaryHike**
- **Description :** Pourcentage d'augmentation salariale reçu par l'employé.

### **26. StandardHours**
- **Description :** Nombre standard d'heures de travail par semaine (valeur fixe `40`).

### **27. StockOptionLevel**
- **Description :** Niveau d'options sur actions détenues par l'employé (échelle de 0 à 3).

### **28. EmployeeCount**
- **Description :** Nombre d'employés dans l'entreprise (valeur fixe `1`).

### **29. EmployeeID**
- **Description :** Identifiant unique de l'employé.
